```python
import numpy as np
from numpy import sqrt,tanh,sinh,cosh,exp,pi,arctanh
from scipy.special import factorial
dtype=np.complex128
# float64 1.7E +/- 308 (15 digits)
```

```python
def LadderOperator(s):
    n=int(2*s+1)
    rtval=np.zeros((n,n))
    for i in range(n-1):
        m=-s+i
        rtval[i+1,i]=sqrt(s*(s+1)-m*(m+1))
    return rtval

print(LadderOperator(2))
print(np.linalg.matrix_power(LadderOperator(2),4))
```

```python
def getTh(th,m12):
    #m12=m1/m2
    value1=m12+cosh(th)+sqrt(1+m12**2+2*m12*cosh(th))
    value2=sinh(th)/value1
    return 4*arctanh(value2)
```

```python


def Amn(m,n,th,th1,th2,C1,C2):
    t=np.arange(m+1)
    k=np.arange(n+1)

    sum1=np.sum(C1[t]*tanh(-th1/4)**(m-t)/(factorial(t)*factorial(m-t)))
    sum2=np.sum(C2[k]*tanh(-th2/4)**(n-k)/(factorial(k)*factorial(n-k)))

    sum3=np.sum(C1[t]*tanh(th1/4)**(m-t)/(factorial(t)*factorial(m-t)))
    sum4=np.sum(C2[k]*tanh(th2/4)**(n-k)/(factorial(k)*factorial(n-k)))

    prod1=exp(2*th)/4*sum1*sum2
    prod2=exp(-2*th)/4*sum3*sum4*(-1)**(m+n)

    return factorial(m+n-1)/sinh(th)*(prod1+prod2)


C1=np.ones(999)
C2=np.ones(999)

print(Amn(m=16,n=16,th=.9,th1=.9,th2=.9,C1=C1,C2=C2))
print(Amn(m=32,n=32,th=.9,th1=.9,th2=.9,C1=C1,C2=C2))
print(Amn(m=64,n=64,th=.9,th1=.9,th2=.9,C1=C1,C2=C2))

```

```python

th=.9
r1=r2=b=1.0
m1=m2=1.0
s1=s2=2

def Chi(s1,s2,m1,m2,r1,r2,b,th,C1,C2):
    b=np.array(b,dtype=dtype)
    th1,th2=getTh(th,m1/m2),getTh(th,m2/m1)
    
    coeff1=-2*pi**2*sqrt(r1*r2*m1*m2)/abs(b)
    sum1=np.zeros(((2*s1+1)*(2*s2+1),)*2)

    J1p,J2p=LadderOperator(s1),LadderOperator(s2)
    J1pm=np.eye(2*s1+1)
    for m in range(2*s1+1):
        J2pn=np.eye(2*s2+1)
        for n in range(2*s2+1):

            coeff2=Amn(m=m,n=n,th=th,th1=th1,th2=th2,C1=C1,C2=C2)
            coeff3=1.0/(m1**m*m2**n*np.abs(b)**(m+n-1))

            mat1=np.kron(J1pm,J2pn)
            mat2=(1j)**(m+n)*mat1.T+(-1j)**(m+n)*mat1
            print(m,n,mat1.shape)

            sum1=sum1+coeff2*coeff3*mat2

            J2pn=J2p@J2pn
        J1pm=J1p@J1pm
    return coeff1*sum1

chi=Chi(s1=16,s2=16,m1=1,m2=1,r1=1,r2=1,b=1,th=.9,C1=C1,C2=C2)
```
